```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

def draw_mutiple_heatmap(datas, vmin=0, vmax=2):
    n = len(datas)
    fig = plt.figure(figsize=(4*n + 2, 4))
    for i in range(n):
        ax = fig.add_subplot(1, n, i+1)
        pcm = ax.pcolormesh(datas[i], cmap=plt.cm.jet, vmin=vmin, vmax=vmax)
        # fig.colorbar(pcm, ax=ax)
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)
    cb = fig.colorbar(pcm, ax=ax, aspect=8)
    cb.ax.tick_params(labelsize=24)
    tick_locator=ticker.MaxNLocator(nbins=int(vmax - vmin))  # colorbar
    cb.locator=tick_locator
    cb.update_ticks()
    fig.tight_layout()  # 
    # plt.subplots_adjust(wspace =0, hspace =0)#
datas = [np.random.rand(365, 96) for i in range(6)]
draw_mutiple_heatmap(datas)
```
